```python
class Person: #คลาสคน
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def introduce(self): #
        return f"สวัสดีครับ ผมชื่อ {self.name} อายุ {self.age} ปี"

class Student(Person): #คลาสนักเรียน มีการสืบทอดมาจากคลาสคน
    def __init__(self,name,age,student_id):
        super().__init__(name,age)
        self.student_id = student_id

    def study(self): #
        return f"{self.name} กำลังเรียนหนังสือ"

    def introduce(self):
        parent_intro = super().introduce()
        return f"{parent_intro} รหัสนักศึกษา : {self.student_id}"

person1 = Person("ชยากร",20)
print(person1.introduce())

student1 = Student("ชยากร",20,"671113555")
print(student1.introduce())
print(student1.study())



#ให้สร้าง class Teccher โดยมี Method ตัดเกรดจากคะแนน 5 ช่อง 3 method
# 1 method 1) method รวมคะแนน 5ช่อง จิตพิสัย กลางภาค ปลายภาค ปฏิบัติ งานที่ได้รับมอบหมาย
# 2 method ตัดเกรด 
# 3 method รายงานผลคะแนนและเกรด

class Teacher(Student):
    def __init__(self,name,age,student_id,point,midterm,finalexam,lab,project):
        super().__init__(self,name,age,student_id)
        self.point = point
        self.midterm = midterm
        self.finalexam = finalexam
        self.lab = lab
        self.project = project
    def grade(self):
        total = self.point + self.midterm + self.finalexam + self.lab + self.project
        if total >= 80:
            return "A"
        elif total >= 70:
            return "B"
        elif total >= 60:
            return "C"
        else:
            return "F"
    def show_info(self):
        print(f"|ชื่อ : {self.name} |รหัสนักศึกษา : {self.student_id} |เกรดที่ได้ : {self.grade()}")
student2 = Teacher("ชยากร","67XXXXX",10,30,30,10,20)
print(student2.show_info())
    
```

```python
class Person: #คลาสคน
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def introduce(self): #
        return f"สวัสดีครับ ผมชื่อ {self.name} อายุ {self.age} ปี"

class Student(Person): #คลาสนักเรียน มีการสืบทอดมาจากคลาสคน
    def __init__(self,name,age,student_id):
        super().__init__(name,age)
        self.student_id = student_id

    def study(self): #
        return f"{self.name} กำลังเรียนหนังสือ"

    def introduce(self):
        parent_intro = super().introduce()
        return f"{parent_intro} รหัสนักศึกษา : {self.student_id}"
    
class Teacher(Person):
        def __init__(self,name,age,subject):
            super().__init__(name,age)
            self.subject = subject
            self.score = [] #list เก็บข้อมูล
        def cal_score(self,s1,s2,s3,s4,s5): #จิตพิสัย กลางภาค ปลายภาคปฏิบัติ งานที่ได้รับมอบหมาย
            self.score = [s1,s2,s3,s4,s5] #listเก็บข้อมูลตามลำดับ
            return sum(self.score)
        
        def grade(self):
            total = sum(self.score)
            if total >=80: return "A"
            elif total >=70: return "B"
            elif total >=60: return "C"
            elif total >=50: return "D"
            else: return "F"

        def report(self,student_name):
            grade = self.grade()
            total = sum(self.score)
            print(f"-----รายงานการสอนวิชา {self.subject}-----")
            print(f"อาจารย์ผู้สอน{self.name}")
            print(f"นักเรียน{student_name}")
            print(f"คะแนนรวม {total}")
            print(f"เกรดที่ได้{grade}")
            print("-"*30)

teacher1 = Teacher("สมชาย",30,"Programing")
total_score = teacher1.cal_score(20,15,10,15,12)
teacher1.report("ชยากร")

person1 = Person("ชยากร",20)
print(person1.introduce())

student1 = Student("ชยากร",20,"671113555")
print(student1.introduce())
print(student1.study())
```

Encapsulation

```python
class BankAccount:
    def __init__(self,account_number,balance):
        self.__account_number = account_number
        self.__balance = balance
    def get_balance(self):
        return self.__balance
    
    def deposit(self, amount):
        if amount >0:
            self.__balance += amount
            print(f"ฝากเงิน {amount} บาท สำเร็จ. ยอดเงินคงเหลือ {self.__balance} บาท")
        else:print(f"จำนวนฝากต้องมากว่า 0 บาท")

    def withdraw(self,amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount 
            print(f"ถอดเงิน {amount} บาท สำเร็จ. ยอดเงินคงเหลือ {self.__balance}")
        else: print("ไม่สามารถอนได้")

my_account = BankAccount("XX-XXXX-9999",500)
my_account.deposit(1000)
my_account.withdraw(1500)
#print({})
#------------------------------------------------------------------------------------

try: 
    print(my_account.__balance)
except AttributeError: #สร้าง/เกิด AttributeError
    print("พิสูจการเข้าถึง__balance โดยตรงไม่ได้")
try:
    my_account.__balance=9999999 
    print(f"ยอดคงเหลือหลังแอบแก้ข้อมูล {my_account.get_balance()}")
except:pass
```

```python
#Homework
class Employee:
    def __init__(self,name,id,salary):
        self.name = name
        self.__id = id
        self.__salary = salary
    def get_salary(self):
        return self.__id,self.__salary
    
    def set_salary(self,new_salary):
        self.new_salary = new_salary
        if new_salary >= 15000:
            self.__salary = new_salary
            print(f"อัปเดทเงินเดือนของ {self.name} สำเร็จ เป็น {self.__salary} บาท")
        else:print("เงินเดือนน้อยกว่าค่าแรงขั้นต่ำ ไม่สามารถอัปเดตได้")
        
employee1 = Employee("ชยากร", "AD67",15000)
employee1.set_salary(14000)
```

Polymorphism

```python
class Shape:
    def draw(self):
        raise NotImplemented("Subclass must implement this method")
    
class Circle(Shape):
    def __init__(self,radius):
        self.radius = radius

    def draw(self):
        return f"กำลังวาดวงกลมที่มีรัศมี {self.radius}"

#Class Rectangle กำลังวาดรูปสี่เหลี่ยม

class Rectangle(Shape):
    def __init__(self,Width,Height):
          self.Width = Width
          self.Height = Height

    def draw(self):
     return f"กำลังวาดรูปสี่เหลี่ยมที่มีความกว้างเท่ากับ {self.Width} และความสูงเท่ากับ {self.Height}"
    
class Square(Shape):
    def __init__(self,side):
        self.side = side
    
    def draw(self):
        return f"กำลังวาดรูปสี่เหลี่ยมจัตุรัสที่มีความกว้างแต่ละด้านเท่ากับ {self.side}"

class Equilateral_triangle(Shape):
    def __init__(self,edge):
        self.edge = edge

    def draw(self):
        return f"กำลังสร้างสามเหลี่ยมด้านเท่าที่มีความกว้างของแต่ละด้านเท่ากับ {self.edge}"

def draw_shapes(shapes):
        for shape in shapes:
            print(shape.draw())

circle = Circle(5)
rectangle = Rectangle(10,20)
square = Square(10)
equilateral_triangle = Equilateral_triangle(12)

shapes_to_draw = [circle,rectangle,square,equilateral_triangle]
draw_shapes(shapes_to_draw)
```
